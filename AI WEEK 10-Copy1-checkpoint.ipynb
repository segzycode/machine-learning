```python
# importing the libraries

import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# importing the dataset into the Training set and Test set

dataset = pd.read_csv('Wine.csv')
dataset.head()
```

```python
dataset.info()
```

```python
X = dataset.iloc[:, 0:-1].values
```

```python
from sklearn.decomposition import PCA

pca = PCA().fit(X)
# pca = PCA(n_components =11)
# X = pca.fit_transform(X)
```

```python
width = 14
height = 7
plt.figure(figsize=(width,  height))



plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o', markersize = 7,)
plt.xticks(range(1, 13))
plt.title('Number of Components')
plt.xlabel('Varriance')
plt.ylabel('PCA')
```

```python
pca = PCA(n_components = 2)
x = pca.fit_transform(X)
```

```python
varriance = pca.explained_variance_ratio_
```

```python
varriance.sum()
```

```python
# chossing the right clusters

from sklearn.cluster import KMeans
wcss = []

for i in range (1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# Visulization

# visualising the Regression Results
width = 14
height = 7
plt.figure(figsize=(width,  height))

plt.plot(range(1,11), wcss, marker='o', markersize = 7,)
plt.title('The Elbow Method')
plt.xlabel('No of Cenriods')
plt.ylabel('wcss')
```

```python
kmeans = KMeans(n_clusters = 4, random_state = 42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans
```

```python
width = 14
height = 7
plt.figure(figsize=(width,  height))

plt.scatter(X[y_kmeans ==0, 0], X[y_kmeans ==0, 1], s =30, c = 'red', label = 'Wine 1')
plt.scatter(X[y_kmeans ==1, 0], X[y_kmeans ==1, 1], s =30, c = 'blue', label = 'Wine 2')
plt.scatter(X[y_kmeans ==2, 0], X[y_kmeans ==2, 1], s =30, c = 'green', label = 'Wine 3')
plt.scatter(X[y_kmeans ==3, 0], X[y_kmeans ==3, 1], s =30, c = 'purple', label = 'Wine 4')
#plt.scatter(X[y_kmeans ==4, 0], X[y_kmeans ==4, 1], s =20, c = 'magenta', label = 'Carefull Poor')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 90, c = 'yellow', label = 'centroids')


plt.title('Clusters Of Customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score K$')
plt.legend()
```
